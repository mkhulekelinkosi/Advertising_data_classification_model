```python
data= data.drop_duplicates()
data.shape
data.columns
# we have no NA or duplicates thus we can move on to data manipulation
```

```python
edu_cols = [col for col in data.columns if 'edu' in col]
purchased_cols = [col for col in data.columns if 'Mnt' in col]
compaign_cols = [col for col in data.columns if 'Accept' in col]
type_of_purchase_cols = [col for col in data.columns if 'Num' in col]
relationship_cols = [col for col in data.columns if 'marital' in col]

new_df = data[['cut_id','Age','Income', 'Kidhome', 'Teenhome', 'Recency',
       'Response', 'Customer_Days'] + edu_cols + purchased_cols + compaign_cols + type_of_purchase_cols + relationship_cols ]

new_df.head()


```

```python
new_df['Education'] = new_df[edu_cols].idxmax(axis=1)
new_df['Education'] =new_df['Education'].str.replace('education_', '')
new_df['Marital_Status'] = new_df[relationship_cols].idxmax(axis=1)
new_df['Marital_Status'] = new_df['Marital_Status'].str.replace('marital_', '')

#new_df['compaign'] = new_df[compaign_cols].idxmax(axis=1)
#new_df['compaign'] = new_df['compaign'].str.replace('Accepted', '')

df = new_df.drop(columns = edu_cols + relationship_cols )
df = df.drop(columns = 'AcceptedCmpOverall')
df['Total_Children'] = df[['Kidhome','Teenhome']].sum(axis=1)

purchase_cols = [col for col in df.columns if 'Num' in col]
df['Total_Purchases'] = df[purchase_cols].sum(axis=1)
df.columns
df.head()
```

```python
sum_table = df[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Response']].sum().to_frame(name='Total').reset_index()
sum_table = sum_table.rename(columns={'index': 'Compaign_type'})
sum_table.head(6)
```

```python

sum_table.plot(kind='bar', x='Compaign_type', y='Total')
plot.xlabel('Compaign_type')
plot.ylabel('Total')
plot.title('Total Children in Household')
plot.show()
```
